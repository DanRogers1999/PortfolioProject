```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
#Read data in
df = pd.read_csv(r'C:\Users\darog\OneDrive\Desktop\Data Analyst Portfolio Project 1\PortfolioPythonProject\movies.csv')
df = df.dropna()
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
#View data
df.head()
```

```python
#Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
#Change datatype of budget column to an int
df['budget'] = df['budget'].astype('int64')
#Change datatype of gross column to an int
df['gross'] = df['gross'].astype('int64')
```

```python
#Create correct year column
df['yearcorrect'] = df['released'].astype(str).str[0:4]
```
